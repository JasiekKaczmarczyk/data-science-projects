## Cleaning and manipulating dataset

### Reading Dataset using Pandas

First I edited file in excel manually in order to remove doubled column indexes

#### Reading edited dataset

```python
dataset = pd.read_excel("Data_Kaggle-Survey-2021_Edit.xlsx", sheet_name="Original_Data")
dataset
```

#### Copying dataset to be able to revert to original dataset quickly if necessary

```python
df = dataset.copy()
df
```

#### Creating unique ID for each participant

```python
df=df.reset_index()
df=df.rename(columns={"index":"ID"})
df
```

### Melting Questions
#### Displaying Columns

```python
print(list(df.columns))
print(len(df.columns))
```

#### Setting id_vars and value_vars to unpivot using melt function

```python
id_vars = df[["ID", "Time from Start to Finish (seconds)"]]
value_vars = df.drop(["ID", "Time from Start to Finish (seconds)"], axis=1)
```

#### Melting

```python
df_melted=df.melt(id_vars=id_vars, value_vars=value_vars, var_name="Question", value_name="Answer")
df_melted
```

### Reading full questions data and joining with dataset

```python
questions = pd.read_excel("Data_Kaggle-Survey-2021_Edit.xlsx", sheet_name="QidToQuestion")
questions
```

```python
questions_until_question_mark = questions.copy()
```

#### Spliting Full_Questions column on '?' or ':' chars, grabbing first element and replacing Full_Question column

```python
questions_until_question_mark["Full_Question"]=questions_until_question_mark["Full_Question"].str.split(pat="[\?:]", expand=True)[0]
questions_until_question_mark
```

#### Joining datasets

```python
df_merged = pd.merge(left=df_melted, right=questions_until_question_mark, how="left", left_on="Question", right_on="Question")
df_merged
```

### Removing NaN values in Answer column

```python
df_notna = df_merged[df_merged["Answer"].notna()]
df_notna
```

```python
print("Len before: "+str(len(df_merged)))
print("Len after: "+str(len(df_notna)))
print("Ratio: " + str(len(df_notna)/len(df_merged)))
```

Nearly 80% of dataset had NaN values in Answer column

### Adding new column with only question number (using regex)

```python
df_q = df_notna.copy()
df_q["Q"] = df_q["Question"].str.extract(pat='(\d+)').astype(int)
df_q
```

#### Checking what are 5 questions with most subquestions

```python
def q_with_most_subquestions(data, count=5):
    # sorting values, grabbing 5 questions with most subquestions and finding their indexes (question numbers)
    q_with_most_subquestions=subquestion_to_question.sort_values(ascending=False).head(count).index

    # grabbing question number and full question
    q_to_fullq=data[["Q", "Full_Question"]]
    # checking if question number is in our list of questions with most subquestions
    q_to_fullq=q_to_fullq[data["Q"].isin(q_with_most_subquestions)]
    # grabbing only Full_Question column and extracting
    question_list=q_to_fullq["Full_Question"].drop_duplicates()
    return list(question_list.sort_values())
```

```python
q_with_most_subquestions(data)
```

It returned 10 values instead of predicted 5, it seems that some questions are slightly altered

#### Checking what are most frequent answers to each question

```python
q_answer=data[["Q", "Full_Question", "Answer"]]
q_answer
```

```python
def frequent_answers(data, nr_results=1):
    frames = []
    # what nr has last question
    last_q = data["Q"].max()


    for n in range(1, last_q+1):
        # grabbing subset where Q = current question, grouping by Full_Question, counting frequency of answer and returning "nr_results" first results
        frames.append(data[data["Q"] == n].groupby("Full_Question")["Answer"].value_counts()[:nr_results].to_frame(name="Count"))
    #concatenating frames and returning DataFrame with results 
    return pd.concat(frames)
```

Checking what are most frequent answers to each question

```python
question_summary=frequent_answers(q_answer)
question_summary
```
