## ML Algorithms Test

```python
from sklearn.datasets import load_iris, load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.datasets import make_regression
import numpy as np
import matplotlib.pyplot as plt
```

### K-Nearest Neighbours

```python
iris = load_iris()
X, y = iris.data, iris.target
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=89)
```

```python
from knn import KNN
```

```python
knn=KNN(3)
knn.fit(X_train, y_train)
predictions=knn.predict(X_test)
```

##### Accuracy

```python
np.sum(predictions == y_test) / len(y_test)
```

### Linear Regression

```python
X, y = make_regression(n_samples=1000, n_features=1, noise=20, random_state=123)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=89)
```

```python
plt.figure(figsize=(10,10))
plt.scatter(X, y)
```

```python
from linear_regression import LinearRegression
```

```python
def RMSE(predicted, actual):
    return np.sqrt(np.mean((predicted-actual)**2))
```

```python
lr=LinearRegression(learning_rate=0.01)
lr.fit(X_train, y_train)
predicted=lr.predict(X_test)
```

##### Root Mean Squared Error

```python
RMSE(predicted, y_test)
```

##### Visualisation

```python
plt.figure(figsize=(10,10))
plt.scatter(X, y)
plt.plot(X, lr.predict(X), c='red')
```

### Logistic Regression

```python
cancer = load_breast_cancer()
X, y = cancer.data, cancer.target
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=89)
```

```python
from logistic_regression import LogisticRegression
```

```python
log_reg=LogisticRegression(learning_rate=0.01)
log_reg.fit(X_train, y_train)
predicted=log_reg.predict(X_test)
```

##### Accuracy

```python
np.sum(predicted == y_test)/len(y_test)
```
